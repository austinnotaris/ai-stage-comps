```python
import pandas as pd
import numpy as np
import sys
sys.path.append("../src")
from yoy_data_builder import *
```

```python
df_CY2018 = pd.read_csv("../data/processed/fundamentals_clean_CY2018.csv")
df_CY2019 = pd.read_csv("../data/processed/fundamentals_clean_CY2019.csv")
df_CY2020 = pd.read_csv("../data/processed/fundamentals_clean_CY2020.csv")
df_CY2021 = pd.read_csv("../data/processed/fundamentals_clean_CY2021.csv")
df_CY2022 = pd.read_csv("../data/processed/fundamentals_clean_CY2022.csv")
df_CY2023 = pd.read_csv("../data/processed/fundamentals_clean_CY2023.csv")
df_CY2024 = pd.read_csv("../data/processed/fundamentals_clean_CY2024.csv")
df_LQ = pd.read_csv("../data/processed/fundamentals_clean_LQ.csv")
```

```python
dfs = [df_CY2018, df_CY2019, df_CY2020, df_CY2021, df_CY2022, df_CY2023, df_CY2024, df_LQ]
dfs = [df.assign(row_index=df.index) for df in dfs]
dfs = [df.assign(dataset_id=i) for i, df in enumerate(dfs)]
df = pd.concat(dfs).sort_values(by=["row_index", "dataset_id"]).reset_index(drop=True)
df = df.drop(columns=["row_index", "dataset_id"])
df = df.rename(columns={"Total Revenue ($USDmm, Historical rate)": "Total Revenue ($USDmm)"})
df = df.rename(columns={"Capital Expenditure ($USDmm, Historical rate)": "Capital Expenditure ($USDmm)"})
df["Total Revenue ($USDmm)"] = pd.to_numeric(df["Total Revenue ($USDmm)"], errors="coerce")
df["Gross Margin %"] = pd.to_numeric(df["Gross Margin %"], errors="coerce")
df["Capital Expenditure ($USDmm)"] = pd.to_numeric(df["Capital Expenditure ($USDmm)"], errors="coerce")
```
