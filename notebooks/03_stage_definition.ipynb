```python
df_tev = pd.read_csv("../data/processed/fundamentals_clean_val.csv")
```

```python
df_tev.head(10)
```

```python
df_tev.dtypes
```

```python
df_tev["Total Enterprise Value ($USDmm, 2018)"] = pd.to_numeric(df_tev["Total Enterprise Value ($USDmm, 2018)"], errors="coerce")
df_tev["Total Enterprise Value ($USDmm, 2019)"] = pd.to_numeric(df_tev["Total Enterprise Value ($USDmm, 2019)"], errors="coerce")
df_tev["Total Enterprise Value ($USDmm, 2020)"] = pd.to_numeric(df_tev["Total Enterprise Value ($USDmm, 2020)"], errors="coerce")
df_tev["Total Enterprise Value ($USDmm, 2021)"] = pd.to_numeric(df_tev["Total Enterprise Value ($USDmm, 2021)"], errors="coerce")
df_tev["Total Enterprise Value ($USDmm, 2022)"] = pd.to_numeric(df_tev["Total Enterprise Value ($USDmm, 2022)"], errors="coerce")
df_tev["Total Enterprise Value ($USDmm, 2023)"] = pd.to_numeric(df_tev["Total Enterprise Value ($USDmm, 2023)"], errors="coerce")
df_tev["Total Enterprise Value ($USDmm, 2024)"] = pd.to_numeric(df_tev["Total Enterprise Value ($USDmm, 2024)"], errors="coerce")
df_tev["Total Enterprise Value ($USDmm, Latest Quarter)"] = pd.to_numeric(df_tev["Total Enterprise Value ($USDmm, Latest Quarter)"], errors="coerce")
df_tev.dtypes
```

```python
df_tev_merge = df_tev.copy()
df_tev_merge.drop(columns="Total Enterprise Value ($USDmm, Latest Quarter)",inplace=True)
cols = df_tev_merge.columns.tolist()
cols[7:] = [col[-5:-1] for col in cols[7:]]
df_tev_merge.columns = cols
df_tev_merge = df_tev_merge.iloc[:,[2,7,8,9,10,11,12,13]]
```

```python
df_tev_merge.head()
```

```python
df_tev_merge = df_tev_merge.melt(id_vars=["Ticker"], var_name="Year", value_name="TEV")
df_tev_merge["Year"] = df_tev_merge["Year"].astype("int64")
```

```python
df_tev_merge.head(24)
```

```python
df_fundamentals = df_fundamentals.merge(df_tev_merge,on=["Ticker","Year"],how="left")
```

```python
df_fundamentals.head(10)
```

```python
df_fundamentals["EV/Rev"] = df_fundamentals["TEV"] / df_fundamentals["Total Revenue ($USDmm)"]
```

```python
df_fundamentals.head(24)
```

```python
df_fundamentals = df_fundamentals[
    [
        "Company Name",
        "Exchange",
        "Ticker",
        "Year",
        "CIK",
        "Excel Company ID",
        "Industry Classifications",
        "Total Revenue ($USDmm)",
        "Revenue YoY",
        "Capex YoY",
        "Gross Margin %",
        "Capital Expenditure ($USDmm)",
        "Capex to Rev",
        "Growth Bucket",
        "Capex Bucket",
        "Margin Bucket",
        "Revenue Bucket",
        "TEV",
        "EV/Rev"
    ]
]
df_fundamentals.head()
```

```python
df_fundamentals.to_csv("../data/processed/full_panel.csv")
```
