```python
def clean(df, time_period):
    df.columns = df.iloc[6]
    df.columns.name = None
    df.drop(df.index[0:7],inplace=True)
    df.reset_index(drop=True,inplace=True)
    df = df.rename(columns={"Security Tickers":"Time Period"})
    df["Time Period"] = time_period
    df["Company Name"] = df["Company Name"].str.split("(").str[0]
    df["Company Name"] = df["Company Name"].str[:-1]
    df.columns = (df.columns.str.replace(r"\s*\[.*?\]\s*", " ", regex=True).str.strip())
    df[["Exchange1", "Ticker1"]] = df["Exchange:Ticker"].str.split(":", n=1, expand=True)
    df.insert(1, "Ticker", df["Ticker1"])
    df.insert(1, "Exchange", df["Exchange1"])
    df = df.drop(columns=["Exchange:Ticker", "Exchange1", "Ticker1"])
    return df
```

```python
df_val_lq = clean(df_val_lq, "Latest Quarter")
df_val_historical = clean(df_val_historical, "N/A")
df_CY2018 = clean(df_CY2018, "2018")
df_CY2019 = clean(df_CY2019, "2019")
df_CY2020 = clean(df_CY2020, "2020")
df_CY2021 = clean(df_CY2021, "2021")
df_CY2022 = clean(df_CY2022, "2022")
df_CY2023 = clean(df_CY2023, "2023")
df_CY2024 = clean(df_CY2024, "2024")
df_LQ = clean(df_LQ, "Latest Quarter")
```
